```python
from openai import OpenAI

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key="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"
)

def ask_llm(messages, model="openai/gpt-4o-mini"):
    response = client.chat.completions.create(
        model=model,
        messages=messages
    )
    return response.choices[0].message.content
```

PERSONAL PROMPT

```python
messages = [
    {"role": "system", "content": "You are an AI professor."},
    {"role": "user", "content": "Explain Machine Learning to a beginner."}
]

print(ask_llm(messages))
```

Cognitive Verifier Pattern

```python
messages = [
    {"role": "user", "content":
     "Explain Blockchain technology. Then verify your explanation for correctness."}
]

print(ask_llm(messages))
```
