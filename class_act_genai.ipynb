AUTO ENCODERS


```python
# Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist

# Step 2: Load MNIST Dataset
(x_train, _), (x_test, _) = mnist.load_data()

# Normalize data (0-1)
x_train = x_train / 255.0
x_test = x_test / 255.0

# Flatten images (28x28 → 784)
x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)

# Step 3: Build Autoencoder Model

input_dim = 784
encoding_dim = 32   # Bottleneck size

# Encoder
input_layer = layers.Input(shape=(input_dim,))
encoder = layers.Dense(128, activation='relu')(input_layer)
encoder = layers.Dense(encoding_dim, activation='relu')(encoder)

# Decoder
decoder = layers.Dense(128, activation='relu')(encoder)
decoder = layers.Dense(input_dim, activation='sigmoid')(decoder)

# Full Autoencoder
autoencoder = models.Model(input_layer, decoder)

# Step 4: Compile Model
autoencoder.compile(optimizer='adam', loss='mse')

# Step 5: Train Model
autoencoder.fit(x_train, x_train,
                epochs=10,
                batch_size=256,
                shuffle=True,
                validation_data=(x_test, x_test))

# Step 6: Reconstruct Images
decoded_images = autoencoder.predict(x_test)

# Step 7: Visualize Results

n = 5  # number of digits to display
plt.figure(figsize=(10,4))

for i in range(n):
    # Original
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_test[i].reshape(28,28), cmap='gray')
    plt.title("Original")
    plt.axis("off")

    # Reconstructed
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(decoded_images[i].reshape(28,28), cmap='gray')
    plt.title("Reconstructed")
    plt.axis("off")

plt.show()
```

An Autoencoder is a type of neural network used for unsupervised learning that learns to compress data into a lower-dimensional representation and then reconstruct it back as close as possible to the original input.

👉 It learns:

Efficient data encoding (compression)

Feature extraction

Noise removal (denoising)

Dimensionality reduction

🔹 Architecture of an Autoencoder

An autoencoder has 3 main parts:

1️⃣ Encoder

Compresses input data into a smaller representation

Learns important features

Output is called Latent Space (Bottleneck layer)

2️⃣ Bottleneck (Latent Space)

The compressed representation

Contains most important information

3️⃣ Decoder

Reconstructs the original input from compressed data

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist

# Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.reshape(-1,28,28,1)/255.0
x_test = x_test.reshape(-1,28,28,1)/255.0

# Correct LeNet Model
lenet = models.Sequential([
    layers.Input(shape=(28,28,1)),   # ✅ Proper Input layer

    layers.Conv2D(6, (5,5), activation='relu'),
    layers.AveragePooling2D(pool_size=(2,2)),  # ✅ pool_size added

    layers.Conv2D(16, (5,5), activation='relu'),
    layers.AveragePooling2D(pool_size=(2,2)),  # ✅ pool_size added

    layers.Flatten(),
    layers.Dense(120, activation='relu'),
    layers.Dense(84, activation='relu'),
    layers.Dense(10, activation='softmax')
])

lenet.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

lenet.summary()

lenet.fit(x_train, y_train, epochs=3)
```

```python
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np

# Load MNIST
(x_train, _), _ = tf.keras.datasets.mnist.load_data()
x_train = x_train / 255.0
x_train = x_train.reshape(-1, 28, 28, 1)

# Generator
def build_generator():
    model = tf.keras.Sequential([
        layers.Dense(128, activation='relu', input_dim=100),
        layers.Dense(784, activation='sigmoid'),
        layers.Reshape((28,28,1))
    ])
    return model

# Discriminator
def build_discriminator():
    model = tf.keras.Sequential([
        layers.Flatten(input_shape=(28,28,1)),
        layers.Dense(128, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    return model

generator = build_generator()
discriminator = build_discriminator()

discriminator.compile(optimizer='adam', loss='binary_crossentropy')

discriminator.trainable = False

gan_input = tf.keras.Input(shape=(100,))
fake_img = generator(gan_input)
gan_output = discriminator(fake_img)

gan = tf.keras.Model(gan_input, gan_output)
gan.compile(optimizer='adam', loss='binary_crossentropy')

print("GAN Model Ready")
```

GAN (Generative Adversarial Network) consists of:

Generator → Creates fake data

Discriminator → Detects real vs fake

They train in competition.

What is U-Net + Implementation (Simple Image Segmentation)
🔹 What is U-Net?

U-Net is a CNN architecture used for image segmentation.
It has:

Encoder (Contracting path) → captures features

Decoder (Expanding path) → reconstructs segmented image

Skip connections → preserve spatial information

```python
# Install dependencies (if needed)
!pip install torch torchvision matplotlib

import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# ---------- Create Synthetic Dataset ----------
def create_circle_image(size=64):
    img = np.zeros((size, size))
    mask = np.zeros((size, size))

    x, y = np.random.randint(15, 50, 2)
    r = np.random.randint(5, 15)

    for i in range(size):
        for j in range(size):
            if (i-x)**2 + (j-y)**2 < r**2:
                img[i,j] = 1
                mask[i,j] = 1

    return img, mask

# Generate dataset
images = []
masks = []

for _ in range(200):
    img, mask = create_circle_image()
    images.append(img)
    masks.append(mask)

images = torch.tensor(images).unsqueeze(1).float()
masks = torch.tensor(masks).unsqueeze(1).float()

# ---------- U-Net Model ----------
class UNet(nn.Module):
    def __init__(self):
        super(UNet, self).__init__()

        self.enc1 = nn.Sequential(
            nn.Conv2d(1, 16, 3, padding=1),
            nn.ReLU(),
            nn.Conv2d(16, 16, 3, padding=1),
            nn.ReLU()
        )

        self.pool = nn.MaxPool2d(2)

        self.enc2 = nn.Sequential(
            nn.Conv2d(16, 32, 3, padding=1),
            nn.ReLU(),
            nn.Conv2d(32, 32, 3, padding=1),
            nn.ReLU()
        )

        self.up = nn.ConvTranspose2d(32, 16, 2, stride=2)

        self.dec = nn.Sequential(
            nn.Conv2d(32, 16, 3, padding=1),
            nn.ReLU(),
            nn.Conv2d(16, 1, 1)
        )

    def forward(self, x):
        e1 = self.enc1(x)
        p = self.pool(e1)
        e2 = self.enc2(p)
        u = self.up(e2)
        cat = torch.cat([u, e1], dim=1)
        out = self.dec(cat)
        return torch.sigmoid(out)

model = UNet()
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# ---------- Training ----------
for epoch in range(10):
    optimizer.zero_grad()
    outputs = model(images)
    loss = criterion(outputs, masks)
    loss.backward()
    optimizer.step()
    print(f"Epoch {epoch+1}, Loss: {loss.item()}")

# ---------- Visualize ----------
with torch.no_grad():
    pred = model(images[:1])

plt.subplot(1,3,1)
plt.title("Input")
plt.imshow(images[0][0], cmap='gray')

plt.subplot(1,3,2)
plt.title("Mask")
plt.imshow(masks[0][0], cmap='gray')

plt.subplot(1,3,3)
plt.title("Prediction")
plt.imshow(pred[0][0], cmap='gray')
plt.show()
```

**Hierarchical RAG Implementation**
 Student FAQ System

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from transformers import pipeline
import numpy as np

# -----------------------
# Sample Knowledge Base
# -----------------------
docs = [
    "U-Net is a convolutional neural network used for biomedical image segmentation.",
    "RAG stands for Retrieval Augmented Generation.",
    "KL divergence measures the difference between two probability distributions.",
    "Stable diffusion is a generative model that creates images from noise.",
    "Autoencoders are neural networks used for representation learning."
]

# -----------------------
# Embedding Model
# -----------------------
embed_model = SentenceTransformer('all-MiniLM-L6-v2')
doc_embeddings = embed_model.encode(docs)

# -----------------------
# Generator Model
# -----------------------
generator = pipeline("text-generation", model="gpt2")

# -----------------------
# Hierarchical RAG Function
# -----------------------
def hierarchical_rag(query):
    query_embedding = embed_model.encode([query])

    similarities = cosine_similarity(query_embedding, doc_embeddings)

    top_k = 2
    top_indices = similarities[0].argsort()[-top_k:][::-1]

    retrieved_context = " ".join([docs[i] for i in top_indices])

    prompt = f"Context: {retrieved_context}\nQuestion: {query}\nAnswer:"

    result = generator(prompt, max_length=120, num_return_sequences=1)

    return result[0]['generated_text']

# -----------------------
# Test
# -----------------------
print(hierarchical_rag("What is U-Net?"))
```

1️⃣ What is RAG?

RAG (Retrieval-Augmented Generation) is a framework that improves Large Language Model (LLM) responses by:

Retrieving relevant documents from a knowledge base

Augmenting the prompt with retrieved content

Generating a final answer using an LLM

Instead of relying only on pre-trained knowledge, RAG uses external data.

2️⃣ What is Hierarchical RAG?

Hierarchical RAG is an advanced version of RAG where retrieval happens in multiple levels (hierarchies).
